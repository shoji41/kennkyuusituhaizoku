```python
import os
import shutil
from google.colab import drive
import cv2 # OpenCVライブラリをインポート

# 1. Googleドライブをマウント
drive.mount('/content/drive')

# --- 設定項目 ---
VIDEO_PATH = '/content/drive/MyDrive/AI_laboratory_course/眼瞼/hisamovie.MOV'  # 元動画のパス
LOCAL_TMP_DIR = '/content/temp_frames'                                       # Colab内の一時フォルダ
ZIP_FILE_NAME = 'cabbage_photos'                                             # 作成するZIPの名前
DRIVE_SAVE_DIR = '/content/drive/MyDrive/AI_laboratory_course/'              # ドライブ上の保存先フォルダ
INTERVAL = 1                                                                # 何フレームごとに保存するか（例:15フレーム=約0.5秒ごと）

# --- 処理開始 ---

# 一時フォルダのクリーンアップ
if os.path.exists(LOCAL_TMP_DIR):
    shutil.rmtree(LOCAL_TMP_DIR)
os.makedirs(LOCAL_TMP_DIR)

# 動画の読み込み
cap = cv2.VideoCapture(VIDEO_PATH)
count = 0
save_count = 0

print("画像を切り出し中...")
while True:
    ret, frame = cap.read()
    if not ret:
        break

    if count % INTERVAL == 0:
        # 画像ファイル名を連番で保存
        file_name = f"img_{save_count:06d}.jpg"
        cv2.imwrite(os.path.join(LOCAL_TMP_DIR, file_name), frame)
        save_count += 1
    count += 1

cap.release()
print(f"{save_count} 枚の画像を切り出しました。")

# フォルダをZIPに固める
print("ZIPファイルを作成中...")
shutil.make_archive(ZIP_FILE_NAME, 'zip', root_dir=LOCAL_TMP_DIR)

# ZIPファイルをGoogleドライブにコピー
print("Googleドライブに転送中...")
shutil.copy(f"{ZIP_FILE_NAME}.zip", os.path.join(DRIVE_SAVE_DIR, f"{ZIP_FILE_NAME}.zip"))

print(f"完了しました！ {DRIVE_SAVE_DIR} に {ZIP_FILE_NAME}.zip が保存されました。")
```
